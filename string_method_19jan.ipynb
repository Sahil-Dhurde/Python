```python
url=input("Enter a url:")
parts = url.split("://")
protocol = parts[0]
rest = parts[1]

subparts = rest.split("/", 1)
domain = subparts[0]

if len(subparts) > 1:
    page = "/"+ subparts[1]
else:
    page = "Not Found"

print("\nSeparated Parts:")
print("Protocol :", protocol)
print("Domain   :", domain)
print("Page     :", page)
```
