```python
a=int(input('enter a value:'))
b=int(input('enter b value:'))
op=input('Enter a Operation:')
def add(a,b):
       c=a+b
       return(c)
def sub(a,b):
       c=a-b
       return(c)
def mul(a,b):
       c=a*b
       return(c)
def div(a,b):
       c=a/b
       return(c)
if op == "+" :
           r=add(a,b)
elif op == "-":
           r=sub(a,b)
elif op == "*":
           r=mul(a,b)
else op == "/":
           r=div(a,b)
print('Result:',r)
```
